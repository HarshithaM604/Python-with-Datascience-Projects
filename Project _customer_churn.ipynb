# F)Random Forest

70:30 ratio

```python
x=df.loc[:,["MonthlyCharges","tenure"]]#2 independent
y=df.loc[:,["Churn"]]#dependent column
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rand_frst=RandomForestClassifier(n_estimators=300)# no of decision tree wit this 75.1 is accuaracyl
```

```python
rand_frst.fit(x_train,y_train)
```

```python
y_pred=rand_frst.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
lst=[200,300,400,500,600,700,800]
trees=[]
accu=[]
for i in lst:
    rand_frst=RandomForestClassifier(n_estimators=300)
    rand_frst.fit(x_train,y_train)
    y_pred=rand_frst.predict(x_test)
    a=accuracy_score(y_pred,y_test)
    accu.append(a)
    trees.append(i)
```
